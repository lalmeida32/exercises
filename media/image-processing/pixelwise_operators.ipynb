```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
```

```python
z = np.arange(256)
print(z) 
```

```python
inv = 255 - z
plt.figure(figsize=(4,4))
plt.plot(z, inv)
```

```python
img1 = imageio.imread(input())
img2 = imageio.imread(input())

img1_invert = 255 - img1
img2_invert = 255 - img2
```

```python
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(222)
plt.imshow(img1_invert, cmap='gray')
plt.axis('off')

plt.subplot(223)
plt.imshow(img2, cmap='gray')
plt.axis('off')


plt.subplot(224)
plt.imshow(img2_invert, cmap='gray')
plt.axis('off')
```

```python
c_scale = 255/np.log2(256)

log = c_scale * np.log2(z+1)
plt.figure(figsize=(4,4))
plt.plot(z, log)
```

```python
img3 = imageio.imread(input())

img3_log = (c_scale * np.log2(img3.astype(np.int32)+1)).astype(np.uint8)
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)
plt.imshow(img3, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.imshow(img3_log, cmap='gray')
plt.axis('off')
```

```python
T = 120
```

```python
aux = np.where(img1 > T)
img1_threshold = np.zeros(img1.shape)
img1_threshold[aux] = 1
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.imshow(img1_threshold, cmap='gray')
plt.axis('off')
```

```python
aux = np.where(img2 > T)
img2_threshold = np.zeros(img2.shape)
img2_threshold[aux] = 1
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.imshow(img2_threshold, cmap='gray')
plt.axis('off')
```

```python
plt.figure(figsize=(4,4))

for y in np.arange(0.25, 4.0, 0.25):
    gamma = (255 * np.power(z / 255, y)).astype(np.uint8)
    plt.plot(z, gamma)
```

```python
y = 2.00
img3_gamma = (255 * np.power(img3 / 255, y)).astype(np.uint8)
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)
plt.imshow(img3, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.imshow(img3_gamma, cmap='gray')
plt.axis('off')
```

```python
k = 0.03
sigmoid = (255/(1+np.exp(-k*(z-127)))).astype(np.uint8)
```

```python
plt.figure(figsize=(4,4))
plt.plot(z, sigmoid)
```

```python
img3_sigmoid = (255/(1+np.exp(-k*(img3.astype(np.int32)-127)))).astype(np.uint8)
```

```python
plt.figure(figsize=(8,4))

plt.subplot(121)
plt.imshow(img3, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.imshow(img3_sigmoid, cmap='gray')
plt.axis('off')
```
