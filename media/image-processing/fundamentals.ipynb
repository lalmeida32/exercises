```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
```

```python
img1 = imageio.imread(input())
img2 = imageio.imread(input())
```

```python
print(type(img1))
print(type(img1[0,0]))

print(img1.shape)
print(img2.shape)
```

```python
plt.imshow(img1, cmap='gray')
plt.colorbar()
```

```python
plt.imshow(img2, cmap='gray')
plt.colorbar()
```

```python
size1 = img1.shape
size2 = img2.shape
assert size1 == size2
```

```python
img_sub = np.zeros(size1, dtype=np.int16)
plt.imshow(img_sub, cmap='gray')
```

```python
for x in range(size1[0]):
    for y in range(size1[1]):
        img_sub[x,y] = np.int16(img1[x,y]) - np.int16(img2[x,y])
```

```python
plt.imshow(img_sub, cmap='gray')
plt.colorbar()
```

```python
imin = np.min(img_sub)
imax = np.max(img_sub)

print('Min value', imin)
print('Max value', imax)
```

```python
# Normalize the image
img_sub_norm = (img_sub - imin) / (imax - imin)

plt.imshow(img_sub_norm, cmap='gray')
plt.colorbar()
```

```python
print(type(img_sub_norm[0,0]))
```

```python
# Convert it back to 8 bit representation
img_sub_norm = (img_sub_norm*255).astype(np.uint8)

plt.imshow(img_sub_norm, cmap='gray')
plt.colorbar()
```

```python
imageio.imwrite(input(), img_sub_norm)
```
